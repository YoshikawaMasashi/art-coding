インポート

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import scipy as sp
import scipy.io.wavfile
import os
%matplotlib inline
```

HRTFデータベース読み出し

```python
hrirs = {}
for dir_path in glob.glob('./hrtfs/elev*'):
    elev = int(dir_path.split("elev")[-1])
    hrir = []
    for azim in range(0, 360, 5):
        if abs(elev) != 90:
            dat_path_r = "{}/R{}e{:03}a.dat".format(dir_path, elev, azim)
            dat_path_l = "{}/L{}e{:03}a.dat".format(dir_path, elev, azim)
        else:
            dat_path_r = "{}/R{}e{:03}a.dat".format(dir_path, elev, 0)
            dat_path_l = "{}/L{}e{:03}a.dat".format(dir_path, elev, 0)
        hrir.append(np.stack([np.loadtxt(dat_path_r), np.loadtxt(dat_path_l)], axis=1))
    hrir = np.stack(hrir, axis=0)
    hrirs[elev] = hrir
```

テストのwavファイル読み出し

```python
_, test = sp.io.wavfile.read("./FFNEnding.wav")
test = test.mean(axis=1)
test /= np.max(np.abs(test))
test = test / 2 ** 15
```

```python
os.makedirs("output", exist_ok=True)
for elev in range(-45, 90, 5):
    print(elev)
    os.makedirs("output/elev_{}".format(elev), exist_ok=True)
    for azim in range(0, 360, 5):
        output_r = np.convolve(test[:1000000], hrirs[elev][azim // 5][:, 0])
        output_l = np.convolve(test[:1000000], hrirs[elev][azim // 5][:, 1])
        output = np.stack([output_r, output_l], axis=1)
        output /= np.max(np.abs(output))
        
        sp.io.wavfile.write("output/elev_{}/a{}.wav".format(elev, azim), 44100, output)
```
